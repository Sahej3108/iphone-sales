```python
data.head()
```

```python
count=data[["Product Name","Number Of Ratings","Star Rating"]]
```

```python
count.head()
```

```python
sorted=count.sort_values(["Number Of Ratings"],ascending=False)
sorted=sorted.head(10)
sorted
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(14,6))
sns.barplot(x='Product Name', y='Number Of Ratings', data=sorted, width=0.6, palette='mako')

plt.title('Top Products by Number Of Ratings', fontsize=14, weight='bold')
plt.xlabel('Product Name')
plt.ylabel('Number Of Ratings')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

__which iphone has highest number od reviews__

```python
reviews_df=data[["Product Name","Number Of Reviews"]]
reviews_df=reviews_df.sort_values(["Number Of Reviews"], ascending=False).head(5)
```

```python
plt.figure(figsize=(18,7))
sns.barplot(x='Product Name', y='Number Of Reviews', data=reviews_df, width=0.5, palette='mako')
plt.xticks(rotation=45, ha='right')
plt.title('Top Products by Number of Reviews')
plt.tight_layout()
plt.show()
```

__relation between the sales price and number of rating__

```python
data.head()
```

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=data,
    x='Sale Price',
    y='Number Of Ratings',
    color='teal',
    alpha=0.6,
    s=70  # point size
)

plt.title('Relation Between Sale Price and Number of Ratings', fontsize=14, weight='bold')
plt.xlabel('Sale Price (â‚¹)', fontsize=12)
plt.ylabel('Number of Ratings', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```

__relation between no. of ratings and discount__

```python
plt.figure(figsize=(10,6))

sns.scatterplot(
    data=data,
    x='Discount Percentage',
    y='Number Of Ratings',
    color='teal',
    alpha=0.6,
    s=70  # point size
)

plt.title('Relation Between Sale Price and Number of Ratings', fontsize=14, weight='bold')
plt.xlabel('Sale Price (â‚¹)', fontsize=12)
plt.ylabel('Number of Ratings', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.4)
plt.tight_layout()
plt.show()
```
